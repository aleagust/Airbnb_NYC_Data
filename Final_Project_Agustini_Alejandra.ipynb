## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.
df.isnull().sum().sort_values()
```

```python
## Impute the values as per the datatype of the columns. The missing values will be replaced using:
# -the median for numerical data
# -the mode for object data type.

for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column].dtype):
        df[column] = df[column].fillna(df[column].median())
    elif df[column].dtype == 'object':
        df[column] = df[column].fillna(df[column].mode().iloc[0])
```

```python
#Check if there are still any missing values:
df.isnull().sum().sort_values()
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
## Display the total number of records in the dataframe after removing the duplicates.
len(df)
```

```python
##Create a graph to show the count of various room types avaliable with Airnb:

#Create the chart:
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4)) 
ax = sns.countplot(x='room_type', data=df, palette='Set2')

#Add details:
plt.title('Distribution of Room Types\n')
plt.yticks(np.arange(0, 55000, step=5000))
plt.xlabel('\nRoom Type')
plt.ylabel('Count')

##Display Chart
plt.show()
```

```python
#Make a bar chat to show which room type adheres to more strict cancellation policy:

#Filter the data:
strict_policy_filter = df[df['cancellation_policy'] == 'strict']
strict_policy_count = strict_policy_filter['room_type'].value_counts()

#Create the chart:
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.barplot(x=strict_policy_count.index, y=strict_policy_count.values, palette='Spectral')

#Add details:
plt.title('Count of Rooms with Strict Policy by Room Type\n')
plt.xlabel('\nRoom Type')
plt.ylabel('Counts')

#Display Chart:
plt.show()

#Calculating the room type that adheres to more strict cancellation policy:
most_strict = strict_policy_count.idxmax()
print('\nThe room type with more strict cancellation policy is:')
print(most_strict)
```

```python
##List the prices by neighbourhood group and also mention which is the most expensive neighborhood group for rentals.

#Group the data by neighbourhood group, calculate the mean price for each group and sort the values:
means = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)

#Create the chart:
sns.set(style="white")
plt.figure(figsize=(8, 4)) 
ax = sns.barplot(x='neighbourhood_group', y='price', data=df, order=means.index, palette='Spectral')

#Add details
ax.set_xticklabels(ax.get_xticklabels(), horizontalalignment='center')
plt.title('Average price per Neighborhood Group\n')
plt.xlabel('\nNeighbourhood Group')
plt.ylabel('Average Price')

#Since the means are very similar, the means values were aggregated to enhance the analysis:
room_types = df['room_type'].unique()
for xtick, label in enumerate(ax.get_xticklabels()):
    ax.text(xtick, means[xtick] - 100, f"Mean: ${means[xtick]:.2f}", 
            ha='center', va='top', fontsize=9)
    
#Display chart:
plt.show()

#Calculating the most expensive neighbourhood group:
print("\nMost expensive neighborhood group to rent from:")
print(most_expensive_neighborhood)
```

```python
##List the top 10 neighbourhoods in the increasing order of their price with the help of a horizontal bar graph. 

#Group the data by neighbourhoods, calculate the mean price for each neighbourhood, sort the data and pick the top 10 more expensive:
top_10_neighbourhoods = df.groupby('neighbourhood')['price'].mean().round(2).sort_values(ascending=False).head(10)

#Create the horizontal bar chart
sns.set(style="white")
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_neighbourhoods.values, y=top_10_neighbourhoods.index, palette='magma')

#Add details
plt.xlim(0, 1150)
plt.title('Top 10 Neighbourhoods with Highest Mean Prices\n')
plt.xlabel('\nMean Price')
plt.ylabel('Neighbourhood')

#Add data labels to the bars:
for index, value in enumerate(top_10_neighbourhoods.values):
    plt.text(value, index, f'${value:.2f}', va='center', size='small')
    
#Display Chart
plt.show()
```

```python
##Make a graph to show the cheapest neighbourhood:

#Group the data by neighbourhoods, calculate the mean price for each neighbourhood, sort the data and pick the 10 chapest:
top_cheap_neighbourhoods = df.groupby('neighbourhood')['price'].mean().round(2).sort_values(ascending=True).head(10)

#Create the horizontal bar chart
sns.set(style="white")
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cheap_neighbourhoods.values, y=top_cheap_neighbourhoods.index, palette='Spectral')

#Add details
plt.xlim(0, 550)
plt.title('Top 10 Cheapest Neighbourhoods\n')
plt.xlabel('\nMean Price')
plt.ylabel('Neighbourhood')

#Add data labels to the bars:
for index, value in enumerate(top_cheap_neighbourhoods.values):
    plt.text(value, index, f'${value:.2f}', va='center', size='small')
    
#Display Chart
plt.show()

#Display the cheapest neighbourhood :
cheapest = top_cheap_neighbourhoods.idxmin()
print('The cheapest neighbourhood is:')
print(cheapest)
```

```python
##List the neighbourhoods which offer short term rentals within 10 days. Illustrate with a bar graph

#As we can see in the result below, there are a total of 224 different neighbourhoods in the data set and a bar chart is not
#enough to display the information to get useful insights. Instead, I used the neighbourhood group clasification to count the
#amount of properties that are offering short term rental.

short_term_rentals = df[df['minimum_nights'] <= 10]
neighborhood_counts = short_term_rentals.groupby('neighbourhood').size().reset_index(name='property_count')
print(neighborhood_counts)
```

```python
#Filter the properties that offers a minimum stay of 10 or less days
short_term_rentals = df[df['minimum_nights'] <= 10]

#Group the filtered data by neighbourhood group
neighborhood_counts = short_term_rentals.groupby('neighbourhood_group').size().reset_index(name='property_count')

#Create the chart
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.barplot(x='neighbourhood_group', y='property_count', data=neighborhood_counts, palette='viridis')

#Add details
plt.title('Number of Properties Offering Short-Term Rentals (10 Days or Less)\n')
plt.xlabel('\nNeighbourhood Group')
plt.ylabel('Number of Properties')

#Display the chart
plt.show()
```

```python
##List the prices with respect to room type using a bar graph and also state your inferences.

#Group the data by room type and calculate the mean price:
price_by_room_type = df.groupby('room_type')['price'].mean().round(2).sort_values(ascending=False)

#Create the bar chart
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
ax = sns.barplot(x=price_by_room_type.index, y=price_by_room_type.values, palette='YlOrBr')

#Add details
ax.set_xticklabels(ax.get_xticklabels(), horizontalalignment='center')
plt.title('Prices by room type\n')
plt.xlabel('\nRoom Type')
plt.ylabel('Mean price\n')

room_types = df['room_type'].unique()
for xtick, label in enumerate(ax.get_xticklabels()):
    ax.text(xtick, price_by_room_type[xtick] - 100, f" ${price_by_room_type[xtick]:.2f}", 
            ha='center', va='top', fontsize=9)

#Display Chart
plt.show()

#Inferences:
most_expensive_room = price_by_room_type.idxmax()
print('The room type with most expensive mean prices is:')
print(most_expensive_room)
cheapest_room = price_by_room_type.idxmin()
print('\nThe room type with the cheapest mean price is:')
print(cheapest_room)
price_difference = round(price_by_room_type.max() - price_by_room_type.min(), 2)
print('\nThe difference between the highest and lowest mean prices across all room types is:')
print( '$', price_difference)
price_difference_percentage = round(price_difference / price_by_room_type.max() *100,2)
print("\nThe percentage difference is:")
print(price_difference_percentage,'%')
```

```python
#Create a pie chart that shows distribution of booked days for each neighborhood group.
#Which neighborhood has the highest booking percentage?

#Group the data and calculate the mean for booked days for each neighbourhood group.
booked_days = df.groupby('neighbourhood_group')['days_booked'].mean().sort_values(ascending=False)

#Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(booked_days, labels=booked_days.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))

#Add details
plt.title('Distribution of Booked Days by Neighborhood Group')
plt.show()

#Display the neighbourhood group with highest booking percentage:
highest_booking_percentage = booked_days.idxmax()
print('The neighborhood group with the highest booking percentage is:')
print(highest_booking_percentage)
```
